```python
from bs4 import BeautifulSoup
import requests
from pymongo import MongoClient
import re

mongo = MongoClient('ec2-52-33-206-219.us-west-2.compute.amazonaws.com',27017)
database = mongo.test.test_db
speeches = []
for index in range(9302,10000):
    print index
    webdata = requests.get('https://storyfunding.daum.net/project/{}'.format(index))
    soup= BeautifulSoup(webdata.text)
    info_funds = soup.find_all('span',attrs = {'class':"info_funds"})
    if len(info_funds) < 1:
        continue
    for infof in  info_funds:
        if not re.search('.+\d+원',infof.get_text().encode('utf-8')):
            continue
        targetmoney = re.search('.+\d+원',infof.get_text().encode('utf-8')).group()   
    
    speech ={}
    totalstr = ""
    descriptions = soup.find_all('p',attrs = {'class': 'desc_prjtop'})
    if len(descriptions) < 2: # 위에 저장된 형식을 만약에 2개 이상 존재하지 않을경우 건너뛴다.
        continue
    for i in descriptions:
        totalstr += i.get_text()

    content = soup.find('div',attrs = {'class':'detail'})
    totalstr += content.text
    speech['content'] = totalstr
    speech['_id'] = index # 아이디를 생성하는데 꺼내쓰기 편하도록
    speech['target_money'] = targetmoney
    speeches.append(speech)
database.insert_many(speeches) # 집어넣는걸 모아놨다가 한번에 넣음
```
